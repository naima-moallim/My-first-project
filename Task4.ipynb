Task1

```python
import random

# Predefined list of words
word_list = ["apple", "banana", "cherry", "date", "elderberry", "fig", "grape"]

def choose_word(word_list):
    return random.choice(word_list)

def display_word(word, guessed_letters):
    return ''.join(letter if letter in guessed_letters else '_' for letter in word)

def hangman_game():
    word_to_guess = choose_word(word_list)
    guessed_letters = set()
    attempts_remaining = 6

    print("Welcome to Hangman!")
    while attempts_remaining > 0:
        print("\nWord to guess: " + display_word(word_to_guess, guessed_letters))
        print("Attempts remaining: " + str(attempts_remaining))
        print("Guessed letters: " + ', '.join(guessed_letters))

        guess = input("Guess a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("Invalid input. Please enter a single letter.")
            continue

        if guess in guessed_letters:
            print("You already guessed that letter.")
            continue

        guessed_letters.add(guess)

        if guess in word_to_guess:
            print("Good guess!")
            if all(letter in guessed_letters for letter in word_to_guess):
                print("\nCongratulations! You guessed the word: " + word_to_guess)
                break
        else:
            print("Sorry, that letter is not in the word.")
            attempts_remaining -= 1

    if attempts_remaining == 0:
        print("\nGame over! The word was: " + word_to_guess)

hangman_game()
```

```python
import json

# Function to load existing student grades from a JSON file
def load_grades_from_file(filename):
    try:
        with open(filename, 'r') as file:
            return json.load(file)
    except FileNotFoundError:
        return {}

# Function to save student grades to a JSON file
def save_grades_to_file(filename, grade_book):
    with open(filename, 'w') as file:
        json.dump(grade_book, file)

# Initialize an empty dictionary to store student grades (load existing data if available)
filename = "student_grades.json"
student_grades = load_grades_from_file(filename)

# Function to input grades for a student and subject
def input_grades():
    student_name = input("Enter student name: ")

    while True:
        subject = input("Enter the subject (or type 'done' to finish for this student): ")
        if subject.lower() == 'done':
            break

        grades_str = input(f"Enter comma-separated grades for {student_name} in {subject} (e.g., 90,85,78): ")
        grades = [int(grade) for grade in grades_str.split(",")]

        if student_name not in student_grades:
            student_grades[student_name] = {}

        student_grades[student_name][subject] = grades

# Function to calculate the average grade for a subject
def calculate_subject_average(grade_book, subject):
    total = 0
    count = 0
    for student_grades in grade_book.values():
        if subject in student_grades:
            total += sum(student_grades[subject])
            count += len(student_grades[subject])

    if count == 0:
        return 0

    return total / count

# Function to display grades and averages for each subject
def display_grades(grade_book):
    for student, subjects in grade_book.items():
        for subject, grades in subjects.items():
            average = calculate_subject_average(grade_book, subject)
            grades_str = ", ".join(map(str, grades))
            print(f"{student} - Subject: {subject}, Grades: {grades_str}. Average: {average:.2f}")

# Main loop for the program
while True:
    print("\nStudent Grade Tracker")
    print("1. Input student grades for a subject")
    print("2. Calculate average grade for a subject")
    print("3. Display all student grades and subject averages")
    print("4. Save and Exit")

    choice = input("What would you like to do? (Enter 1, 2, 3, or 4): ")

    if choice == '1':
        input_grades()
    elif choice == '2':
        subject = input("Enter the subject you want to calculate the average grade for: ")
        average = calculate_subject_average(student_grades, subject)
        print(f"The average grade for {subject} is {average:.2f}")
    elif choice == '3':
        if student_grades:
            display_grades(student_grades)
        else:
            print("No student grades to display.")
    elif choice == '4':
        # Save the data to the JSON file before exiting
        save_grades_to_file(filename, student_grades)
        print("Data saved. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter 1, 2, 3, or 4.")
```
